### Given the vertebrate genome, compare the anchors found via BLAST with different outgroups
As defined in the makefiles, two genes that share the same outgroup gene as best hit are considered.

```python
vert = 'hsa'
outg = ['csa', 'cin']

para_df_blast = {}

for og in outg:
	path = f'/home/lagasso/projects/sinteny/dataset/paralogs/{vert}_{og}.paralogs_from_blast.tsv'
	df = pd.read_csv(path, sep='\t', header=0)
	# drop outgroup gene column
	df = df.iloc[:, :2]
	para_df_blast[og] = df
```

```python
set_csa = set(tuple(sorted([row[0], row[1]])) for index, row in para_df_blast['csa'].iterrows())
set_cin = set(tuple(sorted([row[0], row[1]])) for index, row in para_df_blast['cin'].iterrows())
intersection = set_csa.intersection(set_cin)
print(f'Number of common paralog pairs between csa and cin: {len(intersection)}')
print('Number of unique paralog pairs in csa:', len(set_csa))
print('Number of unique paralog pairs in cin:', len(set_cin))
sd_coeff = len(intersection) / ((len(set_csa) + len(set_cin)) / 2)
print(f'Sorensen-Dice coefficient: {sd_coeff:.4f}')
```
